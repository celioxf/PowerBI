# Power Bi

# What is PowerBI?

PowerBi is a Data Visualisation tool. It has a lot of built-in chart designs with huge customization possibilities as well as huge amount of charts created by the community. PowerBi is very popular in business analytics, specially when it comes to reporting, because it makes it easy for end users to clean data, prepare reports and share it with the whole organisation.

It's important to mention that you need an enterprise account to use all the functionalities of PowerBi. Anyone can download the program (PowerBi Desktop) and test it, but you can't use the PowerBi Service, which allows you to publish the dashboard to your organisation (without the enterprise account, it's only possible to publish the report to the world).

With no further due, we can show some of the functionalities of PowerBi.

# Data Cleaning

As I mentioned in the [previous project](link_to_project), in which I loaded the data into redshift, I left some "dirty data" on purpose, in order to show some functionalities of PowerBi.

So, the first thing we will do is click on transform data, which will open Power Query, the data manipulation tool that comes with PowerBi.

![power_query](imgs/pb_power_query.png)

In our case, we can see that the columns product_category and product_category english contain missing values (the green bar under the column titles is not full). Also, the values in those columns are binded by a underscore character and all the letters are in lowercase.

We can clean it, so that it looks better in our reports.

# Replacing Under Scores and Lowercases

In the "transform" menu in Power Query, we can choose "replace values".

![replace](imgs/pb_replace.png)

Here we can pick which values to look for and which values we want to put in place. In our case, we will replace underscores with a blank space.

![replace_menu](imgs/pb_replace_menu.png)

The modification we just made gets registered at the right side of our screen, which help us to verify what we did and comprehend how we got there. 

![step](imgs/pb_step.png)

We can even change it's characteristics to add some more information (metadata) on what we did.

![step_characateristics](imgs/pb_step_characteristic.png)

![step_changed](imgs/pb_changed_characteristics.png)

Now we can change the first letter of every word to uppercase

![uppercase](imgs/pb_uppercase.png)

# Dealing With Missing Values

We can replace the procedure, to replace the missing values by a explicit category.

![missing](imgs/pb_missing.png)

Now this option is even available for us to filter the column with 

![missing_result](imgs/pb_missing_result.png)

Now we can reproduce the first steps for the "product_category_english" column. However, since this column contains a translation for the name of the product categories, we  can use  conditional column in PowerBi to fill the missing values accordingly.

In the next screen, we create a new column called "product_category_english_clean". To do this, we will look for the values "Nao Identificada" in the column "product_category" (that's the value we created to identify missing values) and replace them with "Unknown", which would be the english translation. For all the other entries that PowerBi does not find, we'll simply use the values in the product_category_english.

![conditional](imgs/pb_conditional.png)

Now our columns look like this and every bar is full, which indicates no missing data.

![final](imgs/pb_final_result.png)

*Note: Cleaning data with PowerBi is a "drag and drop thing". This is useful for users without a technological background to go about their daily tasks. On other hand, this also represents limitations (we can only do what PowerBi allows us to do). A workaround would be to program the transformations with R or Python (the creation of a new column we did above is nothing but a simple "if clause" in any programming language. This is all to say that depending on the cleaning we have to do, it'll be better to run a script inside of PowerBi or even clean the data at the source before importing it*.
